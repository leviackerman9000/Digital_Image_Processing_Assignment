**Task 4: Implement an appropriate technique to reduce the brightness of the provided image effectively.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/DIP lab final rework/DIP_Lab_task_04.jpg")
# the image is in bgr mode, converted to rgb:
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Original    Image')
plt.axis('off')
plt.show()
```

```python
gamma = 2.0 #gamma>1 reduces brightness

normalized = img_rgb/255.0
# Gamma correction works on normalized intensity values.
# This step ensures that the pixel values are scaled to [0, 1]

corrected = np.power(normalized, gamma)  #gamma_correction formula

img_darkened = (corrected * 255).astype(np.uint8)

plt.imshow(img_darkened)
plt.title('Brightness   reduced   Image')
plt.axis('off')
plt.show()
```
