**Task 5: Solve the segmentation problem for the image**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/DIP lab final rework/DIP_Lab_task_05.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# we converted the image to gray as sobel filter works on single channel images
plt.imshow(img_gray, cmap ='gray')
plt.axis('off')
plt.title('Original Image (Grayscale)')
plt.show()
```

*applying sobel filter for edge detection (image has been converted to grayscale as sobel works on single channcel images)*

```python
sobel_x = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)

sobel_combine = np.sqrt(sobel_x**2 + sobel_y**2)  # magnitude formula
sobel_clipped = np.clip(sobel_combine, 0, 255)
sobel_final = np.uint8(sobel_clipped) #converting to uint8 to avoid unexpected behaviour

plt.imshow(sobel_final, cmap='gray')
plt.axis('off')
plt.title('Edge Detected using Sobel Filter')
plt.show()
```
