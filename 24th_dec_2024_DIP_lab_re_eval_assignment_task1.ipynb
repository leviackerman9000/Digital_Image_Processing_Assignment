# Task 1: image dimensions, frequency distribution of intensities, range

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

importing the image into a variable


```python
img = cv2.imread("/content/drive/MyDrive/DIP lab final rework/DIP_Lab_task_01.png")
```

```python
plt.imshow(img)
```

dimension and range calculation

```python
# img.shape returns three values - height, width, number of color channels

print(f"Image Dimensions: {img.shape}")

min_intensity = img.min()
max_intensity = img.max()
print(f"Intensity Range: [{min_intensity}, {max_intensity}]")
```

frequency distribution of the image:


```python
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure()
plt.title("Frequency Distribution of Intensities")
plt.xlabel("I N T E N S I T Y")
plt.ylabel("F R E Q U E N C Y")
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```
