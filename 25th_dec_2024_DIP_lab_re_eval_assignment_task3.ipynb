**Task 3:** **Apply image smoothing with appropriate kernel applied. Identify the noise(if any) before processing.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/DIP lab final rework/DIP_Lab_task_03.png")
```

*We can see salt and pepper noise from the given image (task 03), the white pixels are salt while the black pixels are called pepper. That is why this image has salt and pepper noise.*

```python
plt.imshow(img)
plt.axis('off')
plt.title('Original Image with Salt and Pepper Noise')
plt.show()
```

***We will use Median filter to remove the salt and paper noise as this filter works the best in removing this noise.***

```python
#using median filter
new_img = cv2.medianBlur(img, 5)
plt.imshow(new_img)
plt.axis('off')
plt.title('Noise Removed Image after Median Filter')
plt.show()
```
